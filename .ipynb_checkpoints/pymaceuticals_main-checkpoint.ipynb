### Analysis

- When reviewing the final tumor volumes, Capomulin and Ramicane are comparable in their effectiveness. They are the most effective drugs. Capomulin is significantly more effective than Infubinol and Ceftamine.

- There is a positive correlation between mouse weight and average tumor volume. The correlation value is 0.83. As the mouse weight increases, the average tumor volume increases.

- Based on the Mouse Metadata, it appears that we have kept sample bias to a minimum by using a varied set of mice with different ages, weight, and sex. For example, the male to female ratio was 51% Male mice and 49% female mice. This helps to ensure that our data results are reliable and valid.



```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "Resources/Starter_Code/Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "Resources/Starter_Code/Pymaceuticals/data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
Combined_data = pd.merge(mouse_metadata, study_results, how='right')

# Display the data table for preview
Combined_data = Combined_data[['Mouse ID', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 
                               'Drug Regimen', 'Sex', 'Age_months', 'Weight (g)']]
Combined_data.head()
```

```python
# Checking the number of mice.
mice_count=Combined_data["Mouse ID"].value_counts()
mice_count
```

```python
# step 2 for number of mice
number_of_mice=len(mice_count)
number_of_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.(use subset)

duplicate_mice = Combined_data.loc[Combined_data.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_df = Combined_data.loc[Combined_data["Mouse ID"] == "g989", :]
duplicate_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = Combined_data[Combined_data['Mouse ID'].isin(duplicate_mice)==False]
clean_df.head(8)
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice=clean_df["Mouse ID"].value_counts()
clean_mice
```

```python
#step 2 number of clean mice
clean_number_of_mice=len(clean_mice)
clean_number_of_mice
```

```python
# check the data types in the clean df being used in next section 
clean_df.dtypes
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

# Group by Drug Regimen

summary_df = clean_df.groupby('Drug Regimen')

# Calculate the mean, median, standard deviation, and sem for each drug regimen
tumor_mean = summary_df['Tumor Volume (mm3)'].mean()
tumor_median = summary_df['Tumor Volume (mm3)'].median()
tumor_stdev = summary_df['Tumor Volume (mm3)'].std()
tumor_sem = summary_df['Tumor Volume (mm3)'].sem()

# Create DataFrame to summarize calculations
summary_grouped_df = pd.DataFrame({'Mean': tumor_mean, 'Median': tumor_median,
                                   'Standard Deviation': tumor_stdev, 'SEM': tumor_sem})

summary_grouped_df.head(6)
```

```python
#format 'Mean' and 'Median' summary_grouped_df values to 2 decimal points
summary_grouped_df["Mean"] = summary_grouped_df["Mean"].map("{:.1f}".format)
summary_grouped_df["Median"] = summary_grouped_df["Median"].map("{:.1f}".format)


summary_grouped_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_agg =  clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
# Find max timepoint for each mouse

max_tp_df = pd.DataFrame(clean_df.groupby('Mouse ID')['Timepoint'].max().sort_values()).reset_index().rename(columns={'Timepoint': 'max_timepoint'})
max_tp_df

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
# Merge max timepoint onto clean_df
merged_df = pd.merge(clean_df, max_tp_df, on='Mouse ID')
merged_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
drug_values = []

# Locate each drug from our list 'drugs' in the orginal dataframe
for drug in drugs:
    temp_df = merged_df.loc[merged_df['Drug Regimen'] == drug]

    # Create empty list to fill with tumor vol data (for plotting)
    # Create subset dataframes that narrow down to rows only containing the final timepoint for each regimen (Capomulin, Ramicane, Infubinol, Ceftamin
    final_volume_df = temp_df.loc[temp_df['Timepoint'] == temp_df['max_timepoint']]
    
    # Create a series that consists of all final volume values for each regimen
    values = final_volume_df['Tumor Volume (mm3)']
    drug_values.append(values)

    # Calculate and print the interquartile range (IQR) for each regimen (Capomulin, Ramicane, Infubinol, Ceftamin
    quartiles = values.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'IQR for {drug}: {iqr}')
    
    # Find upper and lower bounds to help identify outliers for each regimen (Capomulin, Ramicane, Infubinol, Ceftamin)
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f'Lower Bound for {drug}: {lower_bound}')
    print(f'Upper Bound for {drug}: {upper_bound}')
    
    # Check for outliers
    outliers_count = (values.loc[(final_volume_df['Tumor Volume (mm3)'] >= upper_bound) | 
                                        (final_volume_df['Tumor Volume (mm3)'] <= lower_bound)]).count()
    print(f'Number of {drug} outliers: {outliers_count}')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

# Edit outliers  
outlier_red_dot = dict(marker='o', markerfacecolor='r', markersize=12, markeredgecolor='black')

# Plot box and whisker 
plt.boxplot(drug_values, flierprops=outlier_red_dot)

# Add title and labels to boxplot
plt.title('Final Tumor Volume by Drug')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xticks([1, 2, 3, 4], drugs)

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

Capomulin_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]

mouseID_df = Capomulin_df.loc[Capomulin_df["Mouse ID"] == "l509",:]
mouseID_df.head()

```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# only Campmulin drug regimen
Only_Capomulin_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]

# Find average tumor volume for each mouse
avg_vol_df = pd.DataFrame(Only_Capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().sort_values()).reset_index().rename(columns= {'Tumor Volume (mm3)': 'Avg_Tumor_Vol'})
avg_vol_df = pd.merge(Only_Capomulin_df, avg_vol_df, on='Mouse ID')
avg_vol_df
```

```python
plt.scatter(avg_vol_df['Weight (g)'],avg_vol_df['Avg_Tumor_Vol'], color="royalblue")
plt.xlabel('Weight (g)')
plt.ylabel('Averag Tumor Volume (mm3)')
plt.title('Mouse Weight Vs. Average Tumor Volume')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen


# Calculate the correlation coefficient between mouse weight and average tumor volume
correlation = st.pearsonr(avg_vol_df['Weight (g)'],avg_vol_df['Avg_Tumor_Vol'])

# Print the answer to above calculation
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

# Calculate linear regression
(slope, intercept,rvalue, pvalue, stderr)=st.linregress(avg_vol_df["Weight (g)"],avg_vol_df["Avg_Tumor_Vol"])
regress_values=avg_vol_df["Weight (g)"]* slope + intercept
line_eq= f"y = {round(slope, 2)} x + {round(intercept, 2)}"

# Plot linear regression on top of scatter plot
plt.scatter(avg_vol_df["Weight (g)"],avg_vol_df["Avg_Tumor_Vol"],color='royalblue')
plt.plot(avg_vol_df["Weight (g)"], regress_values, color='red')
plt.annotate(line_eq,(20,36), fontsize=14)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight vs Tumor Volume for Capomulin")
plt.show()
```
